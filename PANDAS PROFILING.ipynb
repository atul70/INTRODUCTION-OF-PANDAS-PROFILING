## How to Combine MUltiple CSV file using GLOB pacakages

```python
 # 1st Import packages and set the working directory...import glob
    # Use glob to match the pattern ‘csv’

extension = 'csv'
all_filenames= [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#Combine all files in the list and export as CSV


combined_file = pd.concat([pd.read_csv(f) for f in all_filenames])
```

```python
# exporting csv file

combined_file.to_csv('combinedfile.csv',index=False,encoding='utf-8-sig')
```
